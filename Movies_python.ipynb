```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create


# Read in the data

df = pd.read_csv(r'C:\Users\Chuah Wei Jie\Desktop\Lets GO\Project\Movies_Python\movies.csv',encoding='latin-1') # encoding='latin-1' due to utf-8 error
```

```python
# let's look at the data

df.head()
```

```python
# checking if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# data types of columns

df.dtypes
```

```python
# changing data type of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# create correct year column

df['year_corrected'] = df['released'].astype(str).str[0:4] # new column year_corrected added

df.head()
```

```python
df = df.sort_values(by=['gross'],inplace=False,ascending=False) # ascending=False --> for descending order
```

```python
#pd.set_option('display.max_rows',None) # to change the settings to show all rows(entire dataset)
pd.set_option('display.max_rows',50) # limit to 50 rows
```

```python
# checking if there are any duplicates

df['company'].drop_duplicates().sort_values(ascending=False) # returns the distinct 'company' values from the table
```

```python
df.head()
```

```python
# scatter plot with Budget vs Gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
# plot Budget vs Gross Earnings using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# finding correlation

df.corr() # Pearson(default), Kendall, Spearman --> different types of correlation
```

```python
df.corr(method='kendall')
```

```python
df.corr(method='spearman')
```

```python
# visualising Correlation matrix using Pearson Correlation

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# to look at correlation for company vs gross, we have to numerise company

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
# visualising Correlation matrix using Pearson Correlation after numerizing all fields

correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# unstack the matrix

df_numerized.corr().unstack()
```

```python
# sort the unstacked matrix by the correlation value

sorted_pairs = df_numerized.corr().unstack().sort_values(ascending=False)

sorted_pairs
```
